```python
# rules to deal of the missing attributes :
#     1.Get rid of the missing data points --> data.dropna(subset=['RM'])
#     2.Get rid of the whole attributes ---> data.drop('RM',axis=1)
#     3.Set the values to the mean,median or 0
# meadian = data['RM'].median()
# data['RM'].fillna(median)

# This is used to fill the null values with meadian which is given in the strategy
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
imputer.fit(data)
```

```python
imputer.statistics_
```

```python
x=imputer.transform(data)
x
```

```python
data_ = pd.DataFrame(data=x,columns=data.columns)
data_label = data_['MEDV']
data_tr = data_.drop('MEDV',axis=1,inplace=False)
data_tr.info()
```

```python
data_tr.describe()
```

there are three types of objects 
Estimators - it is used to estimate the values in parameters
fit() - This method fits the statagy in the missing values

Transformer - tranform method takes input and returns output based on the learnings from fit(). It also has a convenience function called fit_transform() which fits and then transforms .

Predictors - LinearRegession model is  an example of predictor . fit predict are teo common functions . It also gives score function which will evalute the predictions.



```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([('imputer',SimpleImputer(strategy='median')),('std_median',StandardScaler())])
```

```python
housing=my_pipeline.fit_transform(data_tr)
```

```python
housing.shape
```

```python
housing
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
# model = LinearRegression()
# model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(housing,data_label)
```

```python
some_data = data_tr.iloc[:5]
some_labels = data_label.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
housing_predictions = model.predict(housing)
mse = mean_squared_error(data_label,housing_predictions)
before_rmse = np.sqrt(mse)
```

```python
# rmse
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,data_tr,data_label,scoring='neg_mean_squared_error',cv=10)
rmse=np.sqrt(-scores)
```

```python
# rmse
```

```python
def stats(scores):
    print('scores ',scores)
    print('Mean :',scores.mean())
    print('Standard devation : ',scores.std())
stats(rmse)
```
